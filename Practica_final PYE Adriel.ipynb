```python
import pandas as pd
import matplotlib.pyplot as plt


datos = pd.read_csv('//content/grupo1.csv')


print(datos.head())
print(datos.describe())


plt.hist(datos['Calificacion'], bins=10, color='lightgreen', edgecolor='red')
plt.xlabel('Calificaciones', fontsize=14, color='darkgreen')
plt.ylabel('Frecuencia', fontsize=14, color='darkgreen')
plt.title('Histograma de las Calificaciones', fontsize=16, color='forestgreen')


plt.show()


```

```python
# MODELO DE REGRESION LINEAL
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")
```

```python
# VISUALIZACION RESULTADOS Y CORRELACION
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='red', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='blue', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal de Horas de Estudio v/s Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es:", coef_corr)


```

```python

from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr


horas_estudio = datos[['HorasEstudio']]
calificaciones = datos['Calificacion']


modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)


pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print("Ecuación de la línea de regresión:")
print(f"Calificaciones = {pendiente:.2f} * Horas de Estudio + {intercepto:.2f}")

```

```python
# VISUALIZACION Y COEFICIENTE
import numpy as np
plt.scatter(datos['HorasEstudio'], datos['Calificacion'], color='gold', label='Datos')
plt.plot(datos[['HorasEstudio']], modelo.predict(horas_estudio), color='silver', label='Línea de regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Regresión Lineal Horas de Estudio v/s Calificaciones')
plt.legend()
plt.show()
coef_corr, _ = pearsonr(datos['HorasEstudio'], datos['Calificacion'])
print("El coeficiente de correlación de Pearson es el siguiente:", coef_corr)
```

####¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?
*El coeficiente de regresión en un modelo de regresión lineal representa el cambio esperado en la variable dependiente (calificaciones finales) por cada unidad de cambio en la variable independiente (tiempo de estudio), manteniendo constantes todas las demás variables del modelo..*

####¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?
El coeficiente de correlación es una medida estadística que cuantifica la fuerza y la dirección de la relación lineal entre dos variables. En el contexto de la relación entre el tiempo de estudio y las calificaciones finales, El coeficiente de correlación ofrece una visión clara de cómo dos variables se relacionan en términos de dirección y esfuerzo.
